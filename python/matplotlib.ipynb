```python
x=[2,4,6,8,10,12,14]
y=[8,6,4,4,4,6,8]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
plt.axis((0,20,0,20))
plt.plot(x,y,marker="<",mec="red",mfc="red",ms=8,mew=5,linestyle="-.")
plt.xlabel("Age",color="red",fontsize=20)
plt.ylabel("Age",color="green",fontsize=20)
plt.title("My Graph",color="green",fontsize=20)
plt.show()
```

```python
import numpy as np
```

```python
arr=np.array([1,2,3,4])
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.title("Graph 1")
plt.plot(arr,"r--*",label="2020")
plt.text(2,3,"hello")
plt.legend(title="year",markerscale=1)
plt.subplot(2,1,2)
plt.title("Graph 2")
plt.suptitle("Total Graph")
plt.plot(arr**2,"g:>");
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
