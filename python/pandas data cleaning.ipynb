```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel(r"espn cric score.xlsx",header=3)
```

```python
data.columns=data.iloc[2]
data.columns
data.columns=['AC Voges (AUS)',      '2015-2016',               20,               31,
                      7,             1485,           '269*',            61.87,
                   2667,            55.68,                5,                4,
                      2,              186,                5]
data
```

```python
data["BF"].unique()
```

```python
data[data["BF"]=="-"]
```

```python
data[data["BF"]=="-"].index
```

```python
data.drop(data[data["BF"]=="-"].index,inplace=True)
```

```python
data.reset_index(inplace=True)
data.drop("index",axis=1,inplace=True)
```

```python
data["Country"]=data["Player"].str.split(" ").str[-1].replace("[()]","",regex=True)
```

```python
data["Player"]=data["Player"].str.split("(").str[0]
```

```python
data.head(2)
```

```python
data["Start"]=data["Span"].str.split("-").str[0]
data["End"]=data["Span"].str.split("-").str[0]
```

```python
data.drop("Span",axis=1,inplace=True)
```

```python
data.columns
```

```python
data[['Mat',    'Inns',      'NO',    'Runs',      'HS',
           'Ave',      'BF',      'SR',       100,        50,         0,
            '4s',      '6s']]=data[['Mat',    'Inns',      'NO',    'Runs',      'HS',
           'Ave',      'BF',      'SR',       100,        50,         0,
            '4s',      '6s']].replace("[*+]","",regex=True)
```
